```python
import cv2
import torch
import gc
from ultralytics import YOLO

# 1. Load Model
# Using your specified path for the YOLOv11/v12 best weights
model = YOLO('/content/runs/detect/train/weights/best.pt')

# 2. Setup Paths
video_path = '/content/drive/MyDrive/fyp1/3_car.mp4'
output_path = '/content/drive/MyDrive/fyp1/car3final2.5_Result.mp4'

cap = cv2.VideoCapture(video_path)
fps = int(cap.get(cv2.CAP_PROP_FPS))
w, h = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (w, h))

# 3. Prediction Settings
use_half = True if torch.cuda.is_available() else False

# Using stream=True for memory efficiency during inference
results = model.predict(
    source=video_path,
    half=use_half,
    stream=True,
    agnostic_nms=True,
    conf=0.5
)

# --- TRACKING VARIABLES ---
distraction_frames = 0
empty_frames_count = 0
grace_period_seconds = 0.5
grace_period_frames = int(grace_period_seconds * fps) # 0.5s delay buffer

print(f"Processing... Video FPS: {fps} | Grace Period: {grace_period_frames} frames")

# 4. The Processing Loop
for r in results:
    annotated_frame = r.plot()

    if len(r.boxes) > 0:
        # Distraction detected: Increment timer and reset absence counter
        distraction_frames += 1
        empty_frames_count = 0
    else:
        # No distraction detected in this frame
        empty_frames_count += 1

        # Reset timer only if the person is "gone" for more than 0.5s
        if empty_frames_count > grace_period_frames:
            distraction_frames = 0
        else:
            # Continue counting during grace period to maintain UI smoothness
            if distraction_frames > 0:
                distraction_frames += 1

    # Calculate time based on accumulated frames
    elapsed_video_time = round(distraction_frames / fps, 1)

    # 5. Drawing Logic
    if distraction_frames > 0:
        timer_text = f"Timer: {elapsed_video_time}s"

        # --- TIMER TEXT (Black with White Outline for visibility) ---
        cv2.putText(annotated_frame, timer_text, (50, 150),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 4)
        cv2.putText(annotated_frame, timer_text, (50, 150),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

        # --- RESEARCH-BASED MULTI-STAGE ALERTS ---
        # Stage 3: CRITICAL ALERT (Beyond 2.0s)
        # Probability of crash increases non-linearly/exponentially after this point
        if elapsed_video_time > 2.0:
            cv2.putText(annotated_frame, "!!! DISTRACTED ALERT !!!", (50, 100),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3) # RED

        # Stage 2: SAFETY WARNING (Exactly 2.0s)
        # Threshold where crash risk is at least two times that of normal driving
        elif elapsed_video_time == 2.0:
            cv2.putText(annotated_frame, "WARNING: SAFETY LIMIT", (50, 100),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 255), 2) # YELLOW

        # Stage 1: CAUTION (1.0s to 1.9s)
        # Point where human-machine systems typically start monitoring/detecting glances [cite: 104, 109]
        elif elapsed_video_time >= 1.0:
            cv2.putText(annotated_frame, "CAUTION", (50, 100),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 165, 255), 2) # ORANGE

    out.write(annotated_frame)

    # Memory Management
    del r
    gc.collect()

# 6. Cleanup
cap.release()
out.release()
print(f"✅ Processing Complete! Result saved to: {output_path}")
```
